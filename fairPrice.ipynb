```python
from bs4 import BeautifulSoup
import requests
import MySQLdb
db = MySQLdb.connect(host='localhost', user='root', passwd='GnbweTv]', db='test', charset='utf8')
cursor = db.cursor()
def save_goods(addr, items, classification, page=1):
    if items==[]:
        return
    barcodes = []
    for y in items:
        try:
            barcodes.append({'barcode':re.search('/\d{8,13}-',y.a.get('href')).group(0)[1:-1],'good_name':y.a.get('title').encode('utf-8')})
        except:
            continue
    for bar in barcodes:
        query = """INSERT INTO goods_fairPrice VALUES (%(barcode)d, '%(name)s', '%(last_level)s', NULL, NULL, NULL, NULL)"""\
            %{'barcode':int(bar['barcode']), 'name':bar['good_name'].replace("'", "`"), 'last_level':classification.replace("'", "`")}
        try:
            cursor.execute(query)
            db.commit()
        except:
            print 'error/database'
    try:    
        raw_data = requests.get (addr+'?page='+str(page+1),
                                 headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/37.0.2062.120 Safari/537.36'})
        soup = BeautifulSoup(raw_data.text)
        items = soup.findAll(attrs={'class':'item_text'})
#         print addr+'?page='+str(page+1)
    except: 
        items=[]
    save_goods(addr, items, classification,page+1)
            
```

```python
from bs4 import BeautifulSoup
import os,re
num=0
for root, dirs, files in os.walk('/home/bulat/BestPrice/fair-price.biz/'):
    for file_ in files:
        with open(os.path.join(root,file_)) as f:
            soup = BeautifulSoup(f.read())
        items = soup.findAll(attrs={'class':'item_text'})
        if file_.find('?')==-1:
            save_goods('http://fair-price.biz/'+file_[:file_.find('.html')], items, file_[:file_.find('.html')])

```
